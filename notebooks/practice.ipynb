```python
# ------------------------------------------------------------
# Pr치ctica de NumPy: An치lisis de Distribuciones Probabil칤sticas
# Referencias:
# NumPy Reference Manual: https://numpy.org/doc/stable/reference/index.html
# NumPy User Guide (v1.18): https://numpy.org/doc/1.18/numpy-user.pdf
# ------------------------------------------------------------

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# 1勇 Generaci칩n de datos
# -------------------------------
# np.random.normal y np.random.uniform est치n documentadas en:
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html

np.random.seed(42)  # Reproducibilidad

# Distribuci칩n Normal (media=100, desv칤o=15)
normal = np.random.normal(loc=100, scale=15, size=1000)

# Distribuci칩n Uniforme (entre 50 y 150)
uniforme = np.random.uniform(low=50, high=150, size=1000)

# -------------------------------
# 2勇 C치lculo de estad칤sticos
# -------------------------------
# Funciones utilizadas: np.mean, np.median, np.var, np.std, np.min, np.max
# Documentaci칩n: https://numpy.org/doc/stable/reference/routines.statistics.html

def resumen_estadistico(datos, nombre):
    """Imprime los estad칤sticos descriptivos de un conjunto de datos."""
    print(f"\n游늵 {nombre}")
    print(f"Media: {np.mean(datos):.2f}")
    print(f"Mediana: {np.median(datos):.2f}")
    print(f"Varianza: {np.var(datos):.2f}")
    print(f"Desv칤o est치ndar: {np.std(datos):.2f}")
    print(f"M칤nimo: {np.min(datos):.2f}")
    print(f"M치ximo: {np.max(datos):.2f}")

resumen_estadistico(normal, "Distribuci칩n Normal")
resumen_estadistico(uniforme, "Distribuci칩n Uniforme")

# -------------------------------
# 3勇 Visualizaci칩n
# -------------------------------
# Documentaci칩n de Matplotlib: https://matplotlib.org/stable/api/pyplot_summary.html

plt.figure(figsize=(12, 5))

# Subplot 1: Distribuci칩n Normal
plt.subplot(1, 2, 1)
plt.hist(normal, bins=30, color='skyblue', edgecolor='black', alpha=0.7, label='Normal')
plt.title("Distribuci칩n Normal (풮=100, 픢=15)")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Subplot 2: Distribuci칩n Uniforme
plt.subplot(1, 2, 2)
plt.hist(uniforme, bins=30, color='orange', edgecolor='black', alpha=0.7, label='Uniforme')
plt.title("Distribuci칩n Uniforme (50150)")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# T칤tulo general
plt.suptitle("Comparaci칩n de Distribuciones Probabil칤sticas  NumPy y Matplotlib",
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
